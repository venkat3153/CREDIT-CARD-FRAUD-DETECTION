```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcardd=pd.read_csv('/content/creditcard.csv')
```

```python
creditcardd=creditcardd.dropna()
```

```python
creditcardd['Class'].value_counts()
```

dataset is highly unbalanced

```python
legit=creditcardd[creditcardd.Class==0]
fraud=creditcardd[creditcardd.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditcardd.groupby('Class').mean()
```

under sampling

build a sample dataset containing similar distribution of normal and fradulant transcation

```python
legit_sample=legit.sample(n=52)
```

concatenatiting two dataframes

```python
newd=pd.concat([legit_sample,fraud],axis=0)
```

```python
newd.head()
```

```python
newd.tail()
```

```python
newd['Class'].value_counts()
```

```python
newd.groupby('Class').mean()
```

splitting the data into features and target

```python
x=newd.drop(columns=['Class'],axis=1)
y=newd['Class']
```

split data into training and test data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
print(x_train.shape,x_test.shape)
```

model training

```python

model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

model evaluation

```python
x_train_prediction=model.predict(x_train)
```

```python
training_data_acuracy=accuracy_score(x_train_prediction,y_train)
print(training_data_acuracy)
```

```python
x_test_prediction=model.predict(x_test)
```

```python
test_data_acuracy=accuracy_score(x_test_prediction,y_test)
print(test_data_acuracy)
```
